# 处理result/log.txt文件，并可视化。

```python
import matplotlib.pyplot as plt
```

```python
a=[]
with open('result/log.txt','r') as f:
    for line in f.readlines():
        if('>' in line):
            a.append(line)
```

```python
a[:5]
```

```python
X,Y=[],[]
for x,y in list(map(lambda x:x.replace('>','').replace(';number:',',').strip().split(','),a)):
    X.append(int(x))
    Y.append(int(y))
```

```python
X,Y=[],[]
for x,y in map(lambda x:x.replace('>','').replace(';number:',',').strip().split(','),a):
    X.append(int(x))
    Y.append(int(y))
```

```python
fig, ax = plt.subplots(figsize=(18, 5))
ax.plot(X, Y,'b')
ax.set(xlabel='the length of the step', ylabel='the number of found complete gene',
       title='statics of iteration')
ax.grid()
fig.savefig("statics.png")
plt.show()
```
